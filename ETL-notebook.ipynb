```python
# Read CSV_1
csv_file = "Resources/Suicide_2017.csv"
suicide_2017_df = pd.read_csv(csv_file)
suicide_2017_df.head(10)
```

```python
# Read CSV_2
happiness_2017 = "Resources/Country_happiness_2017.csv"
happiness_2017_df = pd.read_csv(happiness_2017)
happiness_2017_df.head(10)
```

```python
# Merge two data frames together
country_data_df = pd.merge(suicide_2017_df,happiness_2017_df, how='outer', on='Country')
country_data_df.head(10)     
```
